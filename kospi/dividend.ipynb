```python
import numpy as np
import pandas as pd
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../data/kospi/price.csv', index_col='date', parse_dates=True)
sn = pd.read_csv("../../data/kospi/share_number.csv", index_col='date', parse_dates=True)
pm = pd.read_csv("../../data/kospi/price_mod.csv", index_col = 'date', parse_dates = True)

# 배당금 단위(천원)
div_t = pd.read_csv('../../data/kospi/dividend.csv', index_col='date', parse_dates=True)
div_t *= 1000

# 삼성전자 제거
df.drop('A005930', axis = 1, inplace=True)
sn.drop('A005930', axis = 1, inplace=True)
div_t.drop('A005930', axis = 1, inplace=True)
pm.drop(['A005930'], axis=1, inplace=True)
```

```python
# 시가총액
market_cap = df * sn
market_cap.index = market_cap.index + pd.tseries.offsets.MonthEnd(0)
m_cap = market_cap.groupby('date').last()

# 월간 수익률
pm.index = pm.index + pd.tseries.offsets.MonthEnd(0)
m_pm = pm.groupby('date').last()
m_ret = (m_pm.fillna(-1) / m_pm.shift(1)).clip(0, 10) - 1

# 1분기는 전전년, 나머지는 전년도 배당 데이터 사용
div = pd.DataFrame(index=m_ret.index).join(div_t).shift(15)

# 매월 리밸런싱 (but 배당금 데이터 1년에 한번 존재 -> 1년에 한번 리밸런싱하는 꼴)
pm_last = pm.groupby('date').last()
```

```python
# 배당수익률 = 이전 배당금 / 현재 주가
div_r = div / pm_last
div_rank = div_r.rank(axis=1, pct=True, ascending=False)

p1_mask = div_rank <= 0.2
p2_mask = (div_rank > 0.2) & (div_rank <= 0.4)
p3_mask = (div_rank > 0.4) & (div_rank <= 0.6)
p4_mask = (div_rank > 0.6) & (div_rank <= 0.8)
p5_mask = (div_rank > 0.8) & (div_rank <= 1.0)

cap_weight = [(m_cap[i].T / m_cap[i].sum(axis=1)).T for i in [p1_mask, p2_mask, p3_mask, p4_mask, p5_mask]]
```

```python
p1_cr = ((m_ret * cap_weight[0].shift(1)).sum(axis = 1) + 1).cumprod()[15:-1]
p2_cr = ((m_ret * cap_weight[1].shift(1)).sum(axis = 1) + 1).cumprod()[15:-1]
p3_cr = ((m_ret * cap_weight[2].shift(1)).sum(axis = 1) + 1).cumprod()[15:-1]
p4_cr = ((m_ret * cap_weight[3].shift(1)).sum(axis = 1) + 1).cumprod()[15:-1]
p5_cr = ((m_ret * cap_weight[4].shift(1)).sum(axis = 1) + 1).cumprod()[15:-1]

sns.set_style('whitegrid')
sns.lineplot(p1_cr.index, p1_cr, label = 'p1')
sns.lineplot(p2_cr.index, p2_cr, label = 'p2')
sns.lineplot(p3_cr.index, p3_cr, label = 'p3')
sns.lineplot(p4_cr.index, p4_cr, label = 'p4')
sns.lineplot(p5_cr.index, p5_cr, label = 'p5')
plt.show()
```

```python
# 배당성향 (dividend payout ratio)
dpr_t = pd.read_csv('../../data/dpr.csv', index_col='date', parse_dates=True)
dpr = pd.DataFrame(index=m_ret.index).join(dpr_t).shift(15)
dpr.drop('A005930', axis = 1, inplace=True)
```

```python
# 배당성향이 음수이거나 상위 20%인 종목 제외
dpr_mask = (dpr > 0) & (dpr.rank(axis=1, pct=True, ascending=False) > 0.8)

div_rank = div_r[dpr_mask].rank(axis=1, pct=True, ascending=False)

p1_dpr_mask = div_rank <= 0.2
p2_dpr_mask = (div_rank > 0.2) & (div_rank <= 0.4)
p3_dpr_mask = (div_rank > 0.4) & (div_rank <= 0.6)
p4_dpr_mask = (div_rank > 0.6) & (div_rank <= 0.8)
p5_dpr_mask = (div_rank > 0.8) & (div_rank <= 1.0)

dpr_cap_weight = [(m_cap[i].T / m_cap[i].sum(axis=1)).T for i in [p1_dpr_mask, p2_dpr_mask, p3_dpr_mask, p4_dpr_mask, p5_dpr_mask]]
```

```python
p1_dpr_cr = ((m_ret * dpr_cap_weight[0].shift(1)).sum(axis = 1) + 1).cumprod()[15:-1]
p2_dpr_cr = ((m_ret * dpr_cap_weight[1].shift(1)).sum(axis = 1) + 1).cumprod()[15:-1]
p3_dpr_cr = ((m_ret * dpr_cap_weight[2].shift(1)).sum(axis = 1) + 1).cumprod()[15:-1]
p4_dpr_cr = ((m_ret * dpr_cap_weight[3].shift(1)).sum(axis = 1) + 1).cumprod()[15:-1]
p5_dpr_cr = ((m_ret * dpr_cap_weight[4].shift(1)).sum(axis = 1) + 1).cumprod()[15:-1]

sns.set_style('whitegrid')
sns.lineplot(p1_dpr_cr.index, p1_dpr_cr, label = 'p1')
sns.lineplot(p2_dpr_cr.index, p2_dpr_cr, label = 'p2')
sns.lineplot(p3_dpr_cr.index, p3_dpr_cr, label = 'p3')
sns.lineplot(p4_dpr_cr.index, p4_dpr_cr, label = 'p4')
sns.lineplot(p5_dpr_cr.index, p5_dpr_cr, label = 'p5')
plt.show()
```
